# Functionalizing Code

```python
#define function, names BSWrange, reqires input "dataFrameIn"
def BSWrange(dataFrameIn):
    #Indent 4
    return dataFrameIn["housing_median_age"].max() - dataFrameIn["housing_median_age"].min()
```

```python
BSWrange(df_house)
```

Code can be ran on any data set that has "housing_median_age"

```python
def BSWrangeAlt(dataColIn):
    #Indent 4
    return dataColIn.max() - dataColIn.min()
```

```python
BSWrangeAlt(df_house["housing_median_age"])
```

Note function changed. Try changing inputs between BSWrange and BSWrangeAlt and see what happens.
