# Part 3: Loops

```python
#for is command to start a loop
#Has several parts: iterator "i"
#Range of iterator, marked after "in"
#Stuff to repeat, intented after "for"
for i in range(0,10):
    print(i) #Note i is not a string.
```

```python
#Notes: Loops care about iterator and change it every time.  Ranges EXCLUDE last number :( but INCLUDE the first number.
```

```python
#Loops can be nestled.  This is sometimes confusing but important.
for i in range(0,2):
    for n in range(1,2):
        print("The value i of is:",i)
        print("The value n of is:",n)
```
